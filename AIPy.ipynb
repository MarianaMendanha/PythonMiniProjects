### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
# Passo a passo
# passo0: entender desafio
# passo1: importar base de dados
# passo2: preparar a base de dados para a inteligência artificial (pré-processamento)
# passo3: criar modelo de IA -> decidir score de crédito
# passo4: escolher melhor modelo
# passo5: usar IA para fazer previsões

# pacotes: pandas e scikit-learn
import pandas as pd
# import sklearn 

tabela = pd.read_csv("clientes.csv")
display(tabela)
display(tabela.info())
```

```python
# passo2: preparar a base de dados para a inteligência artificial (pré-processamento)
# Label Encoder dá códigos para cada coluna de formato objeto (texto)
# colunas para aplicar encoder: profissao, mix_credito, comportamentp_pagamento

# importar encoder
from sklearn.preprocessing import LabelEncoder
# criar encoder
codificador = LabelEncoder()
# aplicar encoder
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```

```python
y = tabela["score_credito"]
x = tabela.drop(columns=["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# modelos de IA
# Árvore de decisão
# KNN -> Vizinhos próximos (Nearest Neighbors)
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# intancia IAs
modelo_tree = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# treina IAs
modelo_tree.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino) 
```

```python
# testar modelos
previsao_tree = modelo_tree.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_teste, previsao_tree))
print(accuracy_score(y_teste, previsao_knn))
```
